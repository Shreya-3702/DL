```python
import pandas as pd
import numpy as np
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Conv2D,MaxPooling2D,Flatten
```

```python
# Load the MNIST datasets from CSV
mnist_train = pd.read_csv('/content/drive/MyDrive/LP-IV-datasets/MNIST(Ass2&3)/mnist_train.csv')
mnist_test = pd.read_csv('/content/drive/MyDrive/LP-IV-datasets/MNIST(Ass2&3)/mnist_test.csv')
```

```python
mnist_train.head()
```

```python
mnist_test.head()
```

```python
x_train = mnist_train.drop('label',axis=1)
y_train = mnist_train['label']
x_test = mnist_test.drop('label',axis=1)
y_test = mnist_test['label']
```

```python
x_train = np.array(x_train)
y_train = np.array(y_train)
x_test = np.array(x_test)
y_test = np.array(y_test)
```

```python
#normalize
x_train = x_train/255.0
x_test = x_test/255.0
```

```python
x_train.shape
```

```python
model = Sequential([
    Flatten(),
    Dense(128, activation='relu'),
    Dense(64, activation='relu'),
    Dense(10,activation='softmax')
])
```

```python
model.compile(loss="sparse_categorical_crossentropy",optimizer='adam',metrics=['accuracy'])
```

```python
h = model.fit(x_train,y_train,epochs=10,validation_data=(x_test,y_test))
```

```python
predictions = model.predict(x_test)
```

```python
predictions = np.argmax(predictions,axis=1)
```
